```python
import matplotlib.pyplot as plt
import pandas as pd
# Load dataset
ds = "titanic.csv"  # Adjust the path accordingly
df = pd.read_csv(ds)
```

```python
print(df.head())
```

```python
# Plot bar chart for gender distribution
gender_counts = df['Sex'].value_counts()

plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution on the Titanic')
plt.tight_layout()
plt.savefig('/Users/upendra/Desktop/PRODIGY_DS_01/gender_distribution.png')
plt.show()
```

```python
# Plot histogram for age distribution
plt.figure(figsize=(8, 6))
plt.hist(df['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution on the Titanic')
plt.tight_layout()
plt.savefig('/Users/upendra/Desktop/PRODIGY_DS_01/age_distribution.png')
plt.show()
```
